```python
class Cliente:
    # 1. Constructor con atributos privados (doble guion bajo __)
    def __init__(self, nombre: str, apellido: str, genero: str, ocupacion: str):
        self.__nombre = nombre
        self.__apellido = apellido
        self.__genero = genero
        self.__ocupacion = ocupacion

    # =======================
    # 2. GETTERS (obtener datos)
    # =======================

    def get_nombre(self):
        return self.__nombre  # Devuelve el nombre

    def get_apellido(self):
        return self.__apellido  # Devuelve el apellido

    def get_genero(self):
        return self.__genero  # Devuelve el género

    def get_ocupacion(self):
        return self.__ocupacion  # Devuelve la ocupación

    # =======================
    # 3. SETTERS (modificar datos)
    # =======================

    def set_nombre(self, nombre):
        self.__nombre = nombre  # Modifica el nombre

    def set_apellido(self, apellido):
        self.__apellido = apellido  # Modifica el apellido

    def set_genero(self, genero):
        self.__genero = genero  # Modifica el género

    def set_ocupacion(self, ocupacion):
        self.__ocupacion = ocupacion  # Modifica la ocupación

    # =======================
    # 4. Método especial para imprimir el objeto
    # =======================

    def __str__(self):
        return (f"DATOS DEL CLIENTE:\n"
                f"Nombre: {self.__nombre}\n"
                f"Apellido: {self.__apellido}\n"
                f"Género: {self.__genero}\n"
                f"Ocupación: {self.__ocupacion}")


# =======================
# 5. Programa principal
# =======================

if __name__ == '__main__':
    # Crear objeto
    cliente1 = Cliente("Maria", "Paz", "Femenino", "Ingeniera")

    # Mostrar datos iniciales
    print(cliente1)

    # =======================
    # 6. Uso de setters (modificar datos)
    # =======================
    cliente1.set_nombre("Ana")
    cliente1.set_ocupacion("Doctora")

    # =======================
    # 7. Uso de getters (obtener datos)
    # =======================
    print("\nNombre actualizado:", cliente1.get_nombre())
    print("Ocupación actualizada:", cliente1.get_ocupacion())

    # Mostrar datos finales
    print("\nDatos actualizados:")
    print(cliente1)
```
